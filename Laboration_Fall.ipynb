### Simulering

Du kommer nu simulera fallrörelsen med hjälp av Eulers stegmetod för tre olika modeller. Rita figurer för att jämföra de olika modellerna.

```python
dt = 0.01 # s    Tidssteg att använda i simuleringen - variera gärna
N = 300 #        Antal steg att ta
t_i = [dt*n for n in range(N)]
```

1. Använd Eulers stegmetod för att bestämma acceleration, hastighet och sträcka för modellen utan luftmotstånd ($F_f = 0$).

```python
# Skriv din egen kod här (ledning: gör en for-loop som lägger till ett värde i taget till varje vektor - 
# använd till exempel
#     v_none.append(v_none[-1] + dt*a_none[-1])
# där [-1] betyder "sista elementet")
A = 9.82
v = [0]
s = [0]

for i in range(N-1):
    v.append(v[-1] + dt*A)
    s.append(s[-1] + v[-1]*dt)

data2 = pandas.DataFrame({'v': np.array(v).T, 's': np.array(s).T})
display(data2)
plt.plot(t_i, v, label="Hastighet")
plt.plot(t_i, s, label="Sträcka")
plt.scatter(t, h)
plt.legend()
plt.show()
```

2. Använd Eulers stegmetod för att bestämma acceleration, hastighet och sträcka för modellen med linjärt luftmotstånd ($F_f = -bv$). Du kan använda det experimentellt bestämda värdet på gränshastigheten för att bestämma $b$ (Hur stor är $F_f$ när gränshastigheten är uppnådd?).

```python
# Skriv din egen kod här
m = 0.005
b = (A*m)/g
print(b)

a = [9.82]
v = [0]
s = [0]

for i in range(N-1):
    a.append(A - (b*v[-1])/m)
    v.append(v[-1] + dt*a[-1])
    s.append(s[-1] + v[-1]*dt)

data2 = pandas.DataFrame({'a':np.array(a).T, 'v': np.array(v).T, 's': np.array(s).T})
display(data2)
plt.plot(t_i, a, label="Acceleration")
plt.plot(t_i, v, label="Hastighet")
plt.plot(t_i, s, label="Sträcka")
plt.scatter(t, h)
plt.legend()
plt.show()
```

3. Använd Eulers stegmetod för att bestämma acceleration, hastighet och sträcka för modellen med kvadratiskt luftmotstånd ($F_f = -Bv^2$). Du kan använda det experimentellt bestämda värdet på gränshastigheten för att bestämma $B$.
